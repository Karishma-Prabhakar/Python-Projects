```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import datetime
import datetime as datetime
```

```python
df = pd.merge(click, buy, on=['SessionID','ItemID'], how='left' , suffixes=('click','buy'))
print(len(df), len(click) + len(buy))
df.head()
```

```python
# feature 4
# explain your feature: How many items are clicked by the hour (Average)

click['TimeStamp']=pd.to_datetime(click.TimeStamp)
click['Hour']=click.TimeStamp.dt.hour
click['Weekday']=click['TimeStamp'].dt.dayofweek.astype(int)+1
```

```python
click_hour_info = click.groupby(['Hour'])['SessionID'].nunique().reset_index(name='Count_h_c')
click_weekday_info = click.groupby(['Weekday'])['SessionID'].nunique().reset_index(name='Count_w_c')
```

```python
click.head()
```

```python
# feature 5
# explain your feature: How many items are bought by the hour (Average)

buy['TimeStamp']=pd.to_datetime(buy.TimeStamp)
buy['Hour']=buy.TimeStamp.dt.hour
buy['Weekday']=buy['TimeStamp'].dt.dayofweek.astype(int)+1

buy_hour_info = buy.groupby(['Hour'])['SessionID'].nunique().reset_index(name='Count_h_b')
buy_weekday_info = buy.groupby(['Weekday'])['SessionID'].nunique().reset_index(name='Count_w_b')
```

```python
buy.head()
```

```python
# feature 6
# explain your feature: Buy Ratio Averaged for Time (Hours)

result_hours = pd.merge(click_hour_info,buy_hour_info,on='Hour', how='inner')
result_hours['Ratio'] = result_hours['Count_h_b']/result_hours['Count_h_c']
```

```python
result_hours.head()
```

```python
# feature 7
# explain your feature: Buy Ratio Averaged for Time (Weekdays)

result_weekdays = pd.merge(click_weekday_info,buy_weekday_info,on='Weekday', how='inner')
result_weekdays['Ratio'] = result_weekdays['Count_w_b']/result_weekdays['Count_w_c']
```

```python
result_weekdays.head()
```

```python
buy['TimeStamp']=pd.to_datetime(buy.TimeStamp)
buy['Hour']=buy.TimeStamp.dt.hour
buy['Weekday']=buy['TimeStamp'].dt.dayofweek.astype(int)+1

buy_result_info = buy.groupby(['Hour'])['SessionID'].nunique().reset_index(name='Result_h_b')
buy_result_info = buy.groupby(['Weekday'])['SessionID'].nunique().reset_index(name='Result_w_b')
```

```python
buy.head()
```

```python
# feature 8
# explain your feature: Buy Ratio Averaged for Category Number

category=df[['SessionID','Category','statusbuy']]
category_num=category.groupby('SessionID')['Category'].nunique().reset_index(name='Category_num')
session_status=category.groupby('SessionID')['statusbuy'].max().reset_index(name='statusbuy')
category_info=pd.merge(category_num,session_status,on='SessionID')
```

```python
category_info.head()
```

```python
category_buy=category_info.groupby('Category_num')['statusbuy'].sum().reset_index(name='buy')
category_num=category_info.groupby('Category_num')['statusbuy'].count().reset_index(name='session_nums')
category_aggr=pd.merge(category_buy,category_num,on='Category_num')
category_aggr['ratio']=category_aggr.buy/category_aggr.session_nums
```

```python
category_aggr
```

```python
# feature 9
# explain your feature: Buy Ratio Averaged for Session Length

session_length=click.groupby('SessionID')['status'].count().reset_index(name='Length')
session_length=pd.merge(session_status,session_length,on='SessionID')
session_length
```

```python
len_buys=session_length.groupby('Length')['statusbuy'].sum().reset_index(name='buys')
len_num=session_length.groupby('Length')['statusbuy'].count().reset_index(name='len_num')
len_aggr=pd.merge(len_buys,len_num,on='Length')
```

```python
len_aggr['ratio']=len_aggr.buys/len_aggr.len_num
len_aggr
```

```python
# feature 10
# explain your feature:Buy Ratio Averaged for Dwell Time

dwell_max=buy.groupby('SessionID')['TimeStamp'].max().reset_index(name='max')
dwell_min=buy.groupby('SessionID')['TimeStamp'].min().reset_index(name='min')

session_dwell=pd.merge(dwell_max,dwell_min,on='SessionID')
session_dwell['dwell']=(session_dwell['max']-session_dwell['min']).astype('timedelta64[m]')
session_dwell.head()
pd.merge(session_dwell,session_status,on='SessionID')
```

```python
session_dwell=pd.merge(session_dwell,session_status,on='SessionID')
session_dwell.head()
```

```python
# feature 11
# explain your feature: Categories count per session

category_count = df.groupby('SessionID')['Category'].nunique()
df["category_count"] = df["SessionID"].map(lambda x: category_count[x])
```
